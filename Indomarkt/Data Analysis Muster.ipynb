```python
import pandas as pd
from datetime import datetime

datelist = pd.date_range(datetime.today(), periods=100).tolist()
```

```python
import random 
numberlist = []
for i in range(0,100):
    x = random.randint(1,100)
    numberlist.append(x)
```

```python
df = pd.DataFrame()
df['datelist'] = datelist
df['numberlist'] = numberlist
```

```python
import matplotlib.pyplot as plt
import numpy as np

start_date = '2022-10-2'
end_date = '2022-11-2'
df = df.loc[(df['datelist'] > start_date) & (df['datelist'] < end_date) ]

#plt.style.use('fivethirtyeight')
plt.style.use('ggplot')
#plt.xkcd()


fig, ax = plt.subplots()

ax.plot(df['datelist'], df['numberlist'],
         color='#ee9b00', label='Erlöse')

ax.axhline(y=np.nanmean(df['numberlist']), color='red', linestyle='--')

ax.legend()

fig.autofmt_xdate()

ax.set_xlabel("Datum")
ax.set_ylabel("Erlöse")
ax.set_title("Erlöse OnlineShop Indomarkt")

plt.tight_layout()

plt.show()
```
